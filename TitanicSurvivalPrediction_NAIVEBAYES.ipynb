Load Dataset

```python
dataset= pd.read_csv('titanic.csv')
```

Summerise dataset

```python
print(dataset)
print(dataset.head())
print(dataset.shape)
```

Mapping Text Data to Binary Value

```python
# nan_indices =dataset['Gender'].isna().sum()
# print(nan_indices)
income_set=set(dataset['Gender'])
dataset['Gender'] = dataset['Gender'].map({'female':0,'male':1}).astype(int)
print(dataset.head())
```

Segregate dataset into X (Input/IndependentVariable) & Y (Output/dependentVariable)

```python
X=dataset.drop('Survived', axis='columns')
X
```

```python
Y=dataset['Survived']
# Y=dataset.Survived
Y
```

Finding & Removing NA values from our Feature x

```python
X.columns[X.isna().any()]  #is not available
```

```python
X.Age=X.Age.fillna(X.Age.mean())
```

```python
X.Fare=X.Fare.fillna(X.Fare.mean())
```

```python
X.columns[X.isna().any()]
```

Splitting Dataset into Train & Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train, Y_test = train_test_split(X,Y,test_size=0.20,random_state=0)
```

Training

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,Y_train)
```

Predicting, wheather Person Survived or Not

```python
pclassNo=int(input("Enter person's Pclass number: "))
gender=int(input("Enter person's gender 0-female 1-male(0 or 1): "))
age=int(input("Enter person's Age: "))
fare=int(input("Enter person's Fare: "))
person=[[pclassNo,gender,age,fare]]
result=model.predict(person)
print(result)

if result == 1:
  print("Person might be Survived")
else:
   print("Person might not be Survived")
```

Prediction for all Test Data

```python
Y_pred=model.predict(X_test)
print(np.column_stack((Y_pred,Y_test)))
```

Accuracy for our Model

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the model:{0}%".format(accuracy_score(Y_test,Y_pred)*100))
```
